# Task
Generate and analyze three types of chirp signals (linear, quadratic, and logarithmic). This involves defining chirp parameters, generating the signals, plotting their time-domain waveforms with appropriate labels, playing the audio for each chirp, and finally, providing comments on their characteristics based on the visual plots and auditory experience.

## Define Chirp Parameters

### Subtask:
Define the initial frequency (f0), final frequency (f1), duration, and sampling frequency (fs) as provided in the problem statement.


**Reasoning**:
Define the chirp parameters (f0, f1, duration, and fs) as specified in the instructions using a Python code block.



```python
f0 = 500  # Initial frequency in Hz
f1 = 2000 # Final frequency in Hz
duration = 5 # Duration in seconds
fs = 44100 # Sampling frequency in Hz

print(f"Initial frequency (f0): {f0} Hz")
print(f"Final frequency (f1): {f1} Hz")
print(f"Duration: {duration} seconds")
print(f"Sampling frequency (fs): {fs} Hz")
```

## Generate Linear Chirp

### Subtask:
Generate a chirp signal with a linear frequency sweep using the defined parameters.


**Reasoning**:
To generate the linear chirp signal, I need to import the `numpy` library for numerical operations and the `chirp` function from `scipy.signal`. Then, I will create a time vector using `np.linspace` and finally use the `chirp` function with the specified parameters and 'linear' method.



```python
import numpy as np
from scipy.signal import chirp

# Create a time vector
t = np.linspace(0, duration, int(fs * duration), endpoint=False)

# Generate the linear chirp signal
linear_chirp = chirp(t, f0=f0, f1=f1, t1=duration, method='linear')

print(f"Linear chirp signal generated with length: {len(linear_chirp)}")
```

## Generate Quadratic Chirp

### Subtask:
Generate a chirp signal with a quadratic frequency sweep using the defined parameters.


**Reasoning**:
To generate the quadratic chirp, I will use the `chirp` function with the 'quadratic' method and the parameters `t`, `f0`, `f1`, and `duration` (as `t1`), then print its length as a confirmation.



```python
quadratic_chirp = chirp(t, f0=f0, f1=f1, t1=duration, method='quadratic')

print(f"Quadratic chirp signal generated with length: {len(quadratic_chirp)}")
```

## Generate Logarithmic Chirp

### Subtask:
Generate a chirp signal with a logarithmic frequency sweep using the defined parameters.


**Reasoning**:
To generate the logarithmic chirp, I will use the `chirp` function with the 'logarithmic' method and the parameters `t`, `f0`, `f1`, and `duration` (as `t1`), then print its length as a confirmation.



```python
logarithmic_chirp = chirp(t, f0=f0, f1=f1, t1=duration, method='logarithmic')

print(f"Logarithmic chirp signal generated with length: {len(logarithmic_chirp)}")
```

**Reasoning**:
Now that all three chirp signals (linear, quadratic, and logarithmic) have been generated, the next step is to visualize their time-domain waveforms. This involves importing `matplotlib.pyplot` for plotting, creating a figure with subplots for each chirp, and then plotting each chirp signal against the time vector `t`.



```python
import matplotlib.pyplot as plt

# Create a figure with three subplots
plt.figure(figsize=(12, 10))

# Plot Linear Chirp
plt.subplot(3, 1, 1)
plt.plot(t, linear_chirp)
plt.title('Linear Chirp Time Domain Waveform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# Plot Quadratic Chirp
plt.subplot(3, 1, 2)
plt.plot(t, quadratic_chirp)
plt.title('Quadratic Chirp Time Domain Waveform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# Plot Logarithmic Chirp
plt.subplot(3, 1, 3)
plt.plot(t, logarithmic_chirp)
plt.title('Logarithmic Chirp Time Domain Waveform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

**Reasoning**:
Now that the chirp signals are generated and their time-domain waveforms are visualized, the next step is to play the audio for each chirp. This will involve importing `IPython.display` for audio playback and then using `Audio` to play each generated chirp signal.



```python
from IPython.display import Audio

print('Playing Linear Chirp:')
display(Audio(linear_chirp, rate=fs))

print('Playing Quadratic Chirp:')
display(Audio(quadratic_chirp, rate=fs))

print('Playing Logarithmic Chirp:')
display(Audio(logarithmic_chirp, rate=fs))
```

## Analyze Chirp Characteristics

### Observations and Comments:

**Linear Chirp:**
*   **Visual:** The time-domain waveform shows a consistently increasing frequency, with cycles becoming progressively closer together. The amplitude appears uniform throughout.
*   **Auditory:** The linear chirp sounds like a steadily rising tone, with the pitch increasing at a constant rate. It's a very clear and predictable sweep from low to high frequency.

**Quadratic Chirp:**
*   **Visual:** The waveform also shows increasing frequency, but the rate of frequency increase accelerates towards the end. The cycles start spread out and become very tightly packed towards the end of the duration, indicating a faster sweep in higher frequencies.
*   **Auditory:** The quadratic chirp starts with a slower rise in pitch, which then rapidly accelerates, sounding like it rushes towards the higher frequencies much quicker at the end compared to the linear chirp.

**Logarithmic Chirp:**
*   **Visual:** Similar to the quadratic chirp, the frequency increases, but the logarithmic sweep means that the frequency change is more pronounced in the lower frequency range and becomes less discernible (though still increasing) in the higher frequency range. The cycles appear to get denser more rapidly at the beginning and then spread out visually compared to the quadratic sweep where it's the opposite.
*   **Auditory:** The logarithmic chirp sounds like the pitch is rising, but the *perceived* rate of change is more uniform across the frequency range. Our ears perceive pitch logarithmically, so a logarithmic chirp sounds like a more 'even' or 'natural' sweep than a linear or quadratic one, despite the absolute frequency change being faster at lower frequencies.

In summary, while all three chirps sweep from 500 Hz to 2000 Hz, their perceived rates of frequency change and visual characteristics differ significantly due to their underlying mathematical functions, with the logarithmic chirp often sounding the most perceptually smooth to human hearing.
